```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 100)
y = sigmoid(z)

plt.plot(z, y)
plt.xlabel('z')
plt.ylabel('$\sigma(z)$')
plt.title('$\sigma(z) = \\frac{1}{1 + e^{-z}}$')
plt.show()
```

```python
def linear_predictor_sigmoid(x, w, b):
    z = w * x + b
    return sigmoid(z)

def plot_linear_predictor_sigmoid(x, w, b):    
    y = linear_predictor_sigmoid(x, w, b)
    plt.plot(x, y, label=f'w={w},b={b}')

x = np.linspace(-10, 10, 100)
plot_linear_predictor_sigmoid(x, 1, 0)
plot_linear_predictor_sigmoid(x, 5, 0)
plot_linear_predictor_sigmoid(x, 1, 5)

plt.title('$\sigma(x, w, b) = \\frac{1}{1 + e^{-wx+b}}$')
plt.legend()
plt.show()
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 100)
y = sigmoid(z)

fig, ax = plt.subplots()
ax.plot(z, y)

new_value = 1.5
ax.axvline(x=new_value, color='r', linestyle='--')
ax.axhline(y=sigmoid(new_value), color='r', linestyle='--')
plt.yticks([0, 1], ['fail', 'pass'])
plt.title('predict pass with 80% chance')
plt.show()
```
